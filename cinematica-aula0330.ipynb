```
def CalVelocidadeHoraria(vInicial, acel, tempo):
  vh = vInicial + (acel * tempo)
  return vh
```

# Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, des):
  t=2*acel*des
  v=vZero+t
  res = round(math.sqrt(v),2)
  return res
```

##  Testes de Torricelli

```
print("Calculo de Torricelli: ",calTorrricelli(0, 3, 8))
```
